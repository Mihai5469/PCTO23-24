```python
import pandas as pd

file_patch = 'dati_case.csv'

dati_case = pd.read_csv(file_patch)
```

Visualizo contenuto file


```python
dati_case.info()
```

```python
dati_case = dati_case.drop(columns= ['BuildingArea', 'YearBuilt'], axis = 1)
```

```python
dati_case.head()
```

```python
dati_case.columns
```

```python
dati_case = dati_case.dropna(axis=0)
```

```python
dati_case.info()
```

```python
y = dati_case.Price

y.head()
```

```python
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude','Longtitude']

x = dati_case[features]

x.head()
```

Rete neurale

```python
from sklearn.tree import DecisionTreeRegressor

modello = DecisionTreeRegressor(random_state = 1)

modello.fit(x, y)
```

```python
print("Proviamo a fare una predizione per queste 5 case:")
print(x.head())
print("La predizione e:")
print(modello.predict(x.head()))
```

```python
data = {"miei dati": [1, 1.0, 156.0, -37.8079, 144.9934]}

df = pd.DataFrame.from_dict(data, orient='index')

print(modello.predict(df))
```
