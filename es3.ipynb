```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

apro il file csv


```python
file_patch = 'pub.csv'

dati = pd.read_csv(file_patch)
```

```python
dati
```

elimino dati non necesari

```python
dati = dati.drop(columns= ['Radio', 'Newspaper'], axis = 1)
```

```python
dati.head()
```

```python
from sklearn.linear_model import LinearRegression #lib per i grafici
```

```python
X = dati[['TV']]
y = dati['Sales']

regression_model = LinearRegression()

regression_model.fit(X, y)

plt.scatter(X, y, color = "blue", label = 'Dati esservati')
plt.plot(X, regression_model.predict(X), color = "red", linewidth = 2, label ="Regresione lineare")

plt.xlabel("TV")
plt.ylabel("Vendite")
plt.legend()
plt.show()
```

```python
nuovi_dati = pd.DataFrame({'TV' : [249]})

previsioni = regression_model.predict(nuovi_dati)
```
